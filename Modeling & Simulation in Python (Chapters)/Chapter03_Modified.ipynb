 In this simulation, we'll keep track of the number of rare customers arriving at the store. We'll also introduce the concept of iterative modeling.

```python
import random
import matplotlib.pyplot as plt
```

```python
class RareCustomerSimulation:
    def __init__(self, total_customers=0, rare_customers=0):
        self.total_customers = total_customers
        self.rare_customers = rare_customers

    def simulate_customer_arrival(self):
        # Simulate the arrival of a rare customer (1% chance)
        self.total_customers += 1
        if random.random() < 0.01:
            self.rare_customers += 1

    def run_simulation(self, num_iterations):
        for _ in range(num_iterations):
            self.simulate_customer_arrival()

rare_customer_sim = RareCustomerSimulation()
num_simulations = 1000
rare_customer_sim.run_simulation(num_simulations)

total_customers = rare_customer_sim.total_customers
rare_customers = rare_customer_sim.rare_customers
rare_customer_percentage = 100 * (rare_customers / total_customers)

print(f"Total Customers: {total_customers}")
print(f"Rare Customers: {rare_customers}")
print(f"Percentage of Rare Customers: {rare_customer_percentage:.2f}%")
```

```python
data = [total_customers - rare_customers, rare_customers]
labels = ['Regular Customers', 'Rare Customers']

plt.bar(labels, data)
plt.xlabel('Customer Types')
plt.ylabel('Count')
plt.title('Customer Type Distribution')
plt.show()
```
